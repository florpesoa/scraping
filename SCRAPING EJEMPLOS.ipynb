# Ejemplo 2 - Scraping de texto

```python
import PyPDF2

pdfFileObj = open('ProctorTesting.pdf', 'rb')

pdfReader = PyPDF2.PdfFileReader(pdfFileObj)

pageObj = pdfReader.getPage(0)  # primera hoja

print(pageObj.extractText())

# cerrar carpeta
texto_extraido=pdfFileObj.close()

texto_extraido
```

# Ejemplo 3 - Scraping tabla de PDF

```python
ejemplo3=tabula.read_pdf("ejemplo.pdf")
ejemplo3b=ejemplo3[0]
ejemplo3b
```

```python
ejemplo3l= ejemplo3b.dropna() #elimino nan para poder trabajar con los datos primero
```

```python
ejemplo3l.columns = ['col1', 'col2', 'col3', 'col4', 'col5', 'col6'] #transformo las columnas para luego trabajar sus nombres a parte
ejemplo3l.columns = ejemplo3l.columns.str.replace('col4', 'Incertidumbre total')
ejemplo3l.columns = ejemplo3l.columns.str.replace('col6', 'Caudal sólido grueso')
ejemplo3l
```

```python
ejemplo3l[['Fecha', 'Hora', 'Escala Media', 'Caudal líquido']] = ejemplo3l['col1'].str.split(' ', expand=True)
ejemplo3l[['Area', 'Perimetro Sup', 'Perímetro mojado']] = ejemplo3l['col2'].str.split(' ', expand=True)
ejemplo3l[['Vel media', 'Vel máxima']] = ejemplo3l['col3'].str.split(' ', expand=True)
ejemplo3l[['Incertidumbre aleatoria', 'Caudal fino']] = ejemplo3l['col5'].str.split(' ', expand=True)

coleliminar = ['col1', 'col2', 'col3', 'col5']
ejemplo3l=ejemplo3l.drop(columns=coleliminar)
ejemplo3l
```

```python
ejemplo3l[ejemplo3l.columns[-1]] = ejemplo3l.pop(ejemplo3l.columns[0])
```

```python
ejemplo3l[ejemplo3l.columns[-2]] = ejemplo3l.pop(ejemplo3l.columns[0])
ejemplo3l
```
